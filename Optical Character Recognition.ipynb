```python
import easyocr #for easy charcter recognition
import numpy as np
from matplotlib import pyplot as plt
import cv2 # importing opencv
```

```python
IMAGE_PATH = r'C:\Users\hp\Downloads\surf.jpeg'
```

```python
#Extrecting the text
reader = easyocr.Reader(['en'], gpu=False)#reader
result = reader.readtext(IMAGE_PATH)
result
#1st part shows where the text is, 2nd part identifies the text,3rd path shows the accuracy with which it was identified
```

```python
top_left = tuple(result[0][0][0]) #[18,18]
bottom_right = tuple(result[0][0][2]) #[293,145]
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)#reading image
img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5) #laying rectangle
img = cv2.putText(img, text, top_left ,font, .5, (255,255,255),2, cv2.LINE_AA) #putting text above rectangle
plt.imshow(img)
plt.show()
```

```python
#Image with multiple texts
IMAGE_PATH = r'C:\Users\hp\Downloads\sign.png'
```

```python
reader = easyocr.Reader(['en'], gpu=False)#reader
result = reader.readtext(IMAGE_PATH)
result
```

```python
img = cv2.imread(IMAGE_PATH)
for detection in result:
    top_left = tuple(int(val) for val in detection[0][0]) 
    bottom_right = tuple(int(val) for val in detection[0][2])
    text = detection[1]
    font = cv2.FONT_HERSHEY_SIMPLEX
    img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5) #laying rectangle
    img = cv2.putText(img, text, top_left ,font, .3, (255,255,255),2, cv2.LINE_AA) #putting text abo

plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
    
```
